IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

FIRST WE CHECK NUMBER OF TRAINIG EXAMPLES AND 
HERE WE ARE USING GRADIENT DESCENT, SO DEFINE LEARNING RATE AND NUMBER OF ITERATIONS AND
CALCULATING THE ERROR AND UPDATE THE THETA

```python
def linear_regression(x_train, y_train):
    m = len(x_train)

    if m == 0:
        raise ValueError("NO EXAMPLES OF DATASET")
    x_train = np.column_stack((np.ones((m, 1)), x_train))
    theta = np.zeros(x_train.shape[1])
    alpha = 0.01
    num_iterations = 1000
    for _ in range(num_iterations):
        predictions = np.dot(x_train, theta)
        error = predictions - y_train
        gradient = (1/m) * np.dot(x_train.T, error)
        theta -= alpha * gradient
    plt.scatter(x_train[:, 1], y_train, color='blue', label='Training Data')
    plt.plot(x_train[:, 1], np.dot(x_train, theta), color='red', label='Linear Regression')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend()
    plt.show()

    return theta
```

LOADING THE DATASET AND READING IT

```python
dataset_path = r'/content/Medical Price Dataset.csv'
data = pd.read_csv(dataset_path)
```

SPLITTING THE WANTED COLUMNS BY X AND Y TRAIN

```python
x_train = data['bmi'].values 
y_train = data['age'].values 
```

FOR MISSING VALUES AND HANDILNG THEM
AND CONVERTING THEM INTO NUMERIC 
AND PLOTTING THEM USING SCATTER PLOT

```python
if np.isnan(x_train).any() or np.isnan(y_train).any():
    raise ValueError("Dataset contains missing values.")
x_train = pd.to_numeric(x_train, errors='coerce')
y_train = pd.to_numeric(y_train, errors='coerce')
theta = linear_regression(x_train, y_train)
print("Theta parameters:", theta)
```
